```python
import sklearn
import pandas as pd
import numpy as np 
from matplotlib import pyplot 
import matplotlib.pyplot as plt
data = pd.read_csv("googleplaystore_user_reviews.csv")
data.head()





```

```python
df=data.groupby('App')
df.agg([np.mean])
```

```python
data.dropna()
df=data.groupby('App')
```

```python
detail=df.Sentiment_Polarity.agg(['mean'])
result=detail.dropna()
```

```python
dataframe = pd.DataFrame(result)
final=dataframe.reset_index()
final.plot.barh(x='App',y='mean',figsize=(15,300))
```

```python
def classify(mean):
  if(mean>0):
    return "GENUINE"
  elif(mean<0):
    return "FRAUD"
```

```python
final["App Classification"]=final["mean"].apply(classify)
final
```

```python
from csv import writer
from csv import reader
```

```python
text = final["App Classification"]
```

```python
with open('googleplaystore_user_reviews.csv', 'r') as read_obj:
    with open('output_1.csv', 'w', newline='') as write_obj:
        csv_reader = reader(read_obj)
        csv_writer = writer(write_obj)
```

```python
for row in csv_reader:
        
        row.append(text)
        
        csv_writer.writerow(row)
```
